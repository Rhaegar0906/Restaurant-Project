```python
#We need to replace the NAN price range values because there is 6173 NAN values which is too many to simply exclude
meanval = df['price_range'].mean()
df['price_range'].fillna(value=meanval, inplace=True)

#Get stars into a 0-1 scale
df['stars'] = df['stars']/5
df
```

```python
#Creating a dictionary that has the counts of all the different categories found in the dataset
categories_list= []
categories_dict = {}
i = 0
for i in range(len(df)):
  new = df.categories.iloc[i]
  new = list(new.split(","))
  for n,s in enumerate(new):
    new[n] = s.strip()

  old = categories_list
  categories_list =  old + new

for c in range(len(categories_list)):
  categories_dict[categories_list[c]]= categories_list.count(categories_list[c])
#print(categories_dict)

#Sorting and take the top 20 (excluding food and restaurant)
sorted_categories = sorted(categories_dict.items(), key=lambda x:x[1], reverse= True)
sorted_categories
new_cols = []
for i in range(2,22):
  new_cols.append(sorted_categories[i][0])
new_cols
```

```python
def bincol(string):
  tf = [string in i for i in df.categories]
  df[string] = tf
for string in new_cols:
  bincol(string)
  df[string] = df[string].astype(int)
df
```
